Creating a function to access the respective metadata on the folders containing crop images, isolate the unique disease states and then assign to each path, resulting in a dataframe containing disease labels and image paths

```python
def maize_disease_organizer(path):

  #read csv containing metadata and then extract the path and disease class

  df = pd.read_csv(path)

  df_new = df[['path','class']]

  return df_new



```

```python
maize = maize_disease_organizer('/content/drive/MyDrive/maize_data/maize_filtered.csv')
```

Lets convert the two columns into a dictionary, allowing us to mainpulate the strings within the path easier

```python
maize_dict = dict(zip(maize['path'], maize['class']))
```

```python
maize_dict['Healthy/HEALTHY_199.JPEG']
```

Organize the image and path into folders for each disease class.  This method of orgnization is best for the usage of the pytorch dataloader.

```python
main_dir = '/content/drive/MyDrive/maize_data_cleaned/'
```

```python
disease_classes_list = list(maize['class'].unique())
```

```python
#create the separate folders within the folder being used to train the model
for x in disease_classes_list:
  os.mkdir(main_dir + x)
```

```python
# move the image paths into the respective disease class folders

for y in list(maize_dict.keys()):
  try:
    old_dir = '/content/drive/MyDrive/maize_data/maize_images/'
    full_old_dir = old_dir + y.split('/')[-1]

    new_dir = '/content/drive/MyDrive/maize_data_cleaned/' + maize_dict[y] + '/' + y.split('/')[-1]
#/content/drive/MyDrive/maize_data/maize_images
    shutil.move(full_old_dir, new_dir)

  except:
    print ('oops image not found')

```
